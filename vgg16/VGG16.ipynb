## Imports

```python
from tensorflow import keras
from tensorflow.keras.preprocessing import image as image_utils
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import glob, os
from sklearn.metrics import accuracy_score
```

## Load Cell

```python
model = keras.models.load_model('model_name')
```

## Scale the Images

```python
def load_and_scale_image(image_path):
    image = image_utils.load_img(image_path, target_size=(244,244))
    return image
```

## Predictions

```python
os.chdir('../data/Datasets/Test/fake') #will need to switch real to fake so that it can get that testng score as well

y_true=[]
y_pred=[]

label = 0 #need to change the label based on what we are predicintg

for file in glob.glob('*.jpg'):
    image = load_and_scale_image(file)
    image = image_utils.img_to_array(image)
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
    image = preprocess_input(image)
    pred = model.predict(image)
    pred_class = np.argmax(pred)
    
    y_true.append(label)
    y_pred.append(pred_class)
    
print('Accuracy: ', accuracy_score(y_true, y_pred))
```
